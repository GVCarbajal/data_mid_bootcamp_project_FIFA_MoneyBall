```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import os

from sklearn.preprocessing import OneHotEncoder, StandardScaler, MinMaxScaler, Normalizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

import warnings # don't show warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("../data_mid_bootcamp_project_FIFA_MoneyBall/fifa21_male2.csv")
df.columns = [df.columns[col_name].lower().replace(' ','_') for col_name in range(len(df.columns))] # Standarize columns
df = df.set_index('id') # Use player's ID as index
# Filter out unnecesary or redundant columns (e.g., all players are male)
df.drop(['player_photo', 'club_logo', 'flag_photo', 'gender'], axis=1, inplace=True) 
df.sample(10)
```

```python
# Give columns more meaningful names

df.rename(columns = {'ova': 'overall_rating', 
                     'bov': 'best_overall',
                     'bp': 'best_position',
                     'pot': 'potential',
                     'w/f': 'weak_foot',
                     'sm': 'skill_moves',
                     'a/w': 'attacking_work_rate',
                     'd/w': 'defensive_work_rate',
                     'ir': 'international_reputation',
                     'pac': 'pace_diving',
                     'sho': 'shooting_handling',
                     'pas': 'passing_kicking',
                     'dri': 'dribbling_reflexes',
                     'def': 'defending_pace',
                     'phy': 'physical_positioning'
                    }, inplace = True)
```

```python
corr_mat = round(df.corr(), 2) #Extent Of Correlation in our Dataset
corr_mat
```

```python
fig, ax = plt.subplots(figsize=(22,20)) #This will set the Width and Height of our visuals
sns.set_context('poster') #This affects things like the size of the labels, lines, and other elements of the plot
#mask = np.zeros_like(corr_mat)
#mask[np.triu_indices_from(mask)] = True # Mask to hide the repeated half of the matrix
sns.heatmap(corr_mat, xticklabels=corr_mat.columns, yticklabels=corr_mat.columns,cmap='jet')#,mask=mask) # Heatmap declaration 
ax.set_title('Multi-Collinearity of FIFA 2021 Attribute') #Title for Our Graph
#plt.savefig('heatmap.png')
```
